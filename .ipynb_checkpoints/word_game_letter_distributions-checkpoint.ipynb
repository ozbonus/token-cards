# Letter Frequencies in Table-Top Word Games

I've been working on a project that, among other things, involves counting letters in text documents. In the course of working on that the thought came to me to investigate the distributions of letters in some popular table-top word games. I hope the results will prove a least a little enlightening to someone interested designing word games, or even just playing them.

## The Games

Here are the board games we'll take a look at, followed by a link to each game's page on BoardGameGeek which is also where I sourced the data, except for My First Bananagrams, where I needed to request it directly from the manufacturer and they helpfully obliged, and Boggle (Scrabble) Slam where I manually counted the pieces from my own copy of the game.

* **Scrabble** ([BoardGameGeek][1], [Website][2]) is the word game most people are likely to be familiar with and I think needs no extra explanation here. Most of the other games are variations on this original; some minor, others significant.
* **Scrabble Junior** ([BoardGameGeek][3], [Website][4]) is a version of Scrabble designed for children and serves as part game and part teaching tool. The rules are simplified and, as you'll see later, the collection of letters differs as well.
* **Super Scrabble** ([BoardGameGeek][5], [Website][6]) appears at first to simply be a bigger version of basic Scrabble, but actually the distribution of letters has been adjusted such that it is easier to make plurals and pronouns.
* **Upwords** ([BoardGameGeek][7], [Website][8]) differs from Scrabble in that letter tiles may be placed on top of others so that, for example, one might change the word *choose* into *cheese*.
* **Bananagrams** ([BoardGameGeek][9], [Website][10]) works like a free-form version of Scrabble where play is not restricted to a board and each player creates their own grid of words.
* **My First Bananagrams** ([BoardGameGeek][11], [Website][12]) is a version of Bananagrams explicitly designed to be educational and includes combination tiles of common bigrams.
* **Boggle Slam** ([BoardGameGeek][13], [Website][14]), also sold as Scrabble Slam, has little in common with its namesakes. This game begins with one four-letter word and players compete to place new letters on top to make new words. Kind of like Upwords, but restricted to a 4x1 grid. What I find very interesting is that the pieces in this game have different letters on each side which must be chosen between.

## Setting Up

I begin by importing `pandas` to read in and manipulate the raw data, `matplotlib` to handle the plotting, `Counter` from the collections module to tally up letters in text documents. The last three lines set styling options to control the font of the plots and how the dataframes are displayed.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from collections import Counter
from string import punctuation

rcParams['font.family'] = 'monospace'
rcParams['font.size'] = 12
pd.options.display.max_columns = 0
%matplotlib inline
```

## Reading in the Data

The data, stored as `csv` files in the `data/` directory, are easily turned into pandas dataframes. All of the datasets in the cell below share columns for `letter` and `count`. Those for Scrabble and Super Scrabble also have a column for `value`, representing how many points that piece is worth, but that column isn't used in this analysis. The files have each been presorted by `letter`, except for those with blank pieces; such rows are always last.

The data for My First Bananagrams and Boggle (Scrabble) Slam have received some additional processing. Both of these games use pieces that have two letters; 13 out of 39 pieces in My First Bananagrams, and all 55 pieces in Boggle (Scrabble) Slam. Rather than counting each two-letter piece as a unique object, I manually split them apart and increased the single-letter counts accordingly. For example, a My First Bananagrams piece with *ea* was handled by increasing each individual letter count by 1. Files that retain counts for the two-letter pieces are located in the `data/` directory.

```python
scrabble = pd.read_csv('data/scrabble.csv')
scrabble_jr = pd.read_csv('data/scrabble_jr.csv')
super_scrabble = pd.read_csv('data/super_scrabble.csv')
upwords = pd.read_csv('data/upwords_10.csv')
bananagrams = pd.read_csv('data/bananagrams.csv')
my_first_bananagrams = pd.read_csv('data/my_first_bananagrams_split.csv')
boggle_slam_counts = pd.read_csv('data/boggle_slam_counts.csv')

games = [scrabble, scrabble_jr, super_scrabble, upwords, bananagrams, my_first_bananagrams, boggle_slam_counts]
names = ['Scrabble', 'Scrabble\nJunior', 'Super\nScrabble', 'Upwords', 'Bananagrams', 'My First\nBananagrams', 'Boggle\nSlam']

# View a sample dataframe.
scrabble
```

Further poking around involves considering what proportion of a game's pieces each letter is. The function `add_proportion` takes a dataset (formatted like the one above) and returns a new one with a proportion column included. It will also be nice to see how much these proportion differ from a baseline distribution, for which I use Scrabble. The function `add_difference` adds the appropriate column.

```python
def add_proportion(df):
    proportion = df['count'] / df['count'].sum()
    df_p = df.assign(proportion=proportion)
    return df_p

def add_difference(reference_df, df):
    diff = abs(reference_df['proportion'] - df['proportion'])
    df_p_d = df.assign(diff = diff)
    return df_p_d

games_p = [add_proportion(g[:26]) for g in games]
games_p_d = [add_difference(games_p[0], g) for g in games_p]
```

Here are the first few rows of the raw numbers for each game side-by-side along with some descriptive statistics.

```python
concat = pd.concat(
    # Keep the letter column only from the Scrabble dataset.
    [games_p_d[0][['letter', 'proportion', 'diff']]] + [g[['proportion', 'diff']] for g in games_p_d[1:]],
    axis=1,
    keys=[n.replace('\n',' ') for n in names],
    sort=False)

display(concat.head())
display(concat.describe().loc[['count', 'mean', 'std', 'min', 'max']])
```

## Plotting 

The plotting function sorts each dataframe by `proportion` and then `letter`. The color of each bar represents the magnitude of the difference of that letter's proportion from the baseline game, Scrabble

```python
def plot_dist(df_list, names, exp, suptitle):
    
    xmax = max(df['proportion'].max() + 0.01 for df in df_list)
    diff = pd.concat(df['diff'] for df in df_list).values
    norm = plt.Normalize(-0.02, diff.max()*1.8)
    cmap = plt.get_cmap('Purples')
    
    fig, ax = plt.subplots(ncols=len(df_list), figsize=(1.8*len(df_list), 13), facecolor='w', dpi=120)
    plt.subplots_adjust(wspace=0.15)
    for i, g in enumerate(df_list.copy()):
        g.sort_values(by=['proportion', 'letter'], ascending=['False', 'True'], inplace=True)
        normalized = norm(g['diff'].values)
        ax[i].barh(data=g, y='letter', width=g['proportion']**exp, color=cmap(normalized))
        ax[i].set(xlim=(0, xmax), xlabel='', xticklabels='')
        ax[i].set(ylabel='', yticklabels=[f'{e[0]} {e[1]:.2f}' for e in g[['letter','proportion']].values])
        ax[i].tick_params(axis='y', which='both', pad=-30, left=False)
        ax[i].tick_params(axis='x', which='both', bottom=False)
        ax[i].spines['top'].set_visible(False)
        ax[i].spines['right'].set_visible(False)
        ax[i].spines['bottom'].set_visible(False)
        ax[i].spines['left'].set_visible(False)
        
    for i, j in enumerate(ax):
        ax[i].set_title(names[i], loc='left', pad=-20)
    
    if suptitle:
        fig.suptitle(suptitle, y=0.94)
    plt.show()
```

Here I plot the letter distributions twice. Once with bars that reflect each letter's proportion within a game, and again with the bars spanning the entire x-axis to make the ordering differences more apparent.

```python
plot_dist(games_p_d, names, 1, 'Letter Distributions of Words Games\nDarker Colors Represent Difference in Proportion from Scrabble')
```

The plot looks interesting on it's own, but I would like to make one more where the bars extend the breadth of the x-axis in order to focus on the differences. Also, I think it looks cool.

```python
plot_dist(games_p_d, names, 0, 'Letter Distributions of Words Games\nDarker Colors Represent Difference in Proportion from Scrabble')
```

There are some clear patterns along with a few interesting differences. Scrabble, Super Scrabble and Bananagrams are all very similar to each other. This is not surprising given the similar rules and targeted age range. Upwords is a noteworthy deviation, having a noticably flatter distribution. Scrabble Junior and My First Bananagrams, the two games with educational goals, are yet more interesting, differing greatly from the "adult" games and from each other. It seems these games were designed with different opinions on what sort of words young children ought to be spelling. 

Boggle Slam sticks out as being quite different from the others. I can only suppose that its particular distribution of letters is better-suited to the game's play style. The next section will investigate the game further.

## Boggle Slam

To restate, Boggle Slam (sold as Scrabble Slam in some places) is also a game about spelling words, but it differs from the other games in two important respects: first, the goal of the game involves spelling words of exactly four letters and, second, the game pieces have one letter on each side, only one of which can be in play. Similar to Upwords, words can be made by placing letters on top of others. Indeed during a typical game most words are made by changing one or two letters of an existing word.

I must wonder how the distribution of letters in the game compares to other bodies of four-letter words. To that end I have built a miniature corpus of two documents. The first is a list a common short words compiled by Josh Kaufman ([Github](https://github.com/first20hours/google-10000-english)) and the other is the full text from Lucy Maud Montgomery's *Anne of Green Gables* ([Project Gutenberg](http://www.gutenberg.org/ebooks/45)), chosen for being a literary work with a low reading level. My version of the book was taken from Shibamouli Lahiri's Gutenberg Dataset ([Website](https://web.eecs.umich.edu/~lahiri/gutenberg_dataset.html)) which has removed the Project Gutenberg Licenses from its documents.

The next cell transforms these two documents into dataframes with the same format as the others so far in this notebook. The code to do this is a bit of mess given the very different structures of the documents, but I hope that it is still readable. The file `stopwords.text` is a list of very common English words that are typically excluded during text analysis. The original list was sourced from the Natural Language Toolkit ([nltk.org](http://www.nltk.org/)) and modified to exlcuded Roman numerals and the name Anne for obvious reasons. Finally the relative frequencies of the letters are plotted with a new colormap pegged to the Boggle Slam distribution.

```python
# An explicitly named variable is more readible.
slam_counts_p = games_p[6]

with open('data/stopwords.txt') as stopwords_file:
    stopwords = stopwords_file.read().split()

mini_corpus = ['data/google-10000-english-usa-no-swears-short.txt',
               'data/Lucy Maud Montgomery___Anne Of Green Gables.txt']

# A list to hold the dataframes until their needed.
four_letter = []

for c in mini_corpus:
    
    counter = Counter()
    
    with open(c, 'r') as file:
        text = file.read().lower().split()
        # Get four-letter words without apostrophes, not found in stopwords, and without punctuation.
    words = [w for w in text
             if len(w)==4
             and w not in stopwords
             and not bool(set(w).intersection(set(punctuation)))]
    counter.update(l for w in words for l in w)
    
    df = pd.DataFrame.from_dict(counter, orient='index')
    df.reset_index(inplace=True)
    df.columns = ['letter', 'count']
    df['proportion'] = df['count'] / df['count'].sum()
    df.sort_values(by=['letter'], inplace=True)
    df.reset_index(inplace=True, drop=True)
    four_letter.append(df)

dfs_p_d = [add_difference(slam_counts_p, df) for df in [slam_counts_p] + four_letter]
names_2 = ['Boggle\nSlam', 'Common Four-\nLetter Words', 'Four-Letter\nWords in AoGG']
plot_dist(dfs_p_d, names_2, 1, 'Proportion of Letters in Boggle (Scrabble)\nSlam Compared to Four-Letter Words')
```

Well, so much for my hypothesis, because those couldn't look much more different! However they arrived at their distribution, it must have been more sophisticated than tallying up a bunch of four-letter words. Like any good game, they must have gone through many hours of playtesting before finding a distribution that worked for them.

## The Other Side of the Card

I have in mind a sort of word game which is similar to Boggle Slam in that the game pieces will have letters on each side. The purpose of that is simply to get more pieces into the players' hands efficiently. As I work on it I wonder how to pair letters together and how the developers of this game made their decision.

Here is the entire deck of 55 cards.

```python
slam_cards = pd.read_csv('data/boggle_slam_cards.csv')
slam_cards.T
```

Here I create two columns that hold the proportion for the letters on either side of the cards, then one more column that is the sum of those two.

```python
slam_cards = slam_cards[:52]
proportion_map = dict(zip(slam_counts_p['letter'], slam_counts_p['proportion']))

slam_cards = slam_cards.assign(obv_proportion=slam_cards['obverse'].map(proportion_map))
slam_cards = slam_cards.assign(rev_proportion=slam_cards['reverse'].map(proportion_map))
slam_cards = slam_cards.assign(sum_proportion=slam_cards['obv_proportion'] + slam_cards['rev_proportion'])
```

Without much to go on I can only throw everything at the wall and see what sticks. To that end, I examine the means, standard deviations, and do some plotting. I import the Seaborn plotting library here in order to access it's pleasant-looking and informative KDE plot.

```python
display(slam_cards.head())
display(slam_cards.describe().loc[['mean', 'std', 'min', 'max']])

import seaborn as sns
fig, ax = plt.subplots(ncols=3, figsize=(15,4))
sns.distplot(slam_cards['obv_proportion'], ax=ax[0])
sns.distplot(slam_cards['rev_proportion'], ax=ax[1])
sns.distplot(slam_cards['sum_proportion'], ax=ax[2])
plt.show()
```

From my own experience playing the game I am operating under the assumption that cards that are more common in the deck are more useful in creating words and less common cards more challenging. Even just a glance at the cards would be enough to arrive at the same conclusion, I think. Given that, looking at `sum_proportion` we can see that the game's designers decided on a set of cards where a large majority could be categorized as "pretty good", having a sum proportion somewhere between 0.08 to 0.12. Even the lowest scoring cards share a sum proportion of 0.06. While these are more difficult to use, getting one of these five cards will not leave a player in a hopeless position.
